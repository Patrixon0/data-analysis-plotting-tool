```python
#

import numpy as np
import matplotlib.pyplot as plt

# Deine Datenpunkte
x_1 = np.array([404.66, 435.83, 467.81, 479.99, 508.58, 546.08, 576.96, 579.06, 643.85])
y_1 = np.array([47.5 + 15/60, 48 + 10/60, 48 + 12/60, 48.5 + 3/60, 49, 49 + 26/60, 49.5 + 10/60, 50 + 26/60, 51 + 18/60])
# y_1 muss gedreht werden
y_1 = [y_1[len(y_1)-i -1] for i in range(len(y_1))]
n = len(x_1)
steigungen = np.zeros(n)

# Lokale Steigungen berechnen
for i in range(n):
    if i == 0:
        # Erste Steigung: nur zum nächsten Punkt
        m1 = (y_1[i+1] - y_1[i]) / (x_1[i+1] - x_1[i])
        steigungen[i] = m1
    elif i == n-1:
        # Letzte Steigung: nur zum vorherigen Punkt
        m1 = (y_1[i] - y_1[i-1]) / (x_1[i] - x_1[i-1])
        steigungen[i] = m1
    else:
        # Normale Punkte: Gewichtete Mittelung der Nachbarsteigungen
        m1 = (y_1[i] - y_1[i-1]) / (x_1[i] - x_1[i-1])
        m2 = (y_1[i+1] - y_1[i]) / (x_1[i+1] - x_1[i])

        d1 = np.sqrt((x_1[i] - x_1[i-1])**2 + (y_1[i] - y_1[i-1])**2)
        d2 = np.sqrt((x_1[i+1] - x_1[i])**2 + (y_1[i+1] - y_1[i])**2)

        steigungen[i] = (m1 * d2 + m2 * d1) / (d1 + d2)

# Schnittpunkte der benachbarten Geraden berechnen
schnittpunkte_x = []
schnittpunkte_y = []

for i in range(n-1):
    m1, m2 = steigungen[i], steigungen[i+1]
    x1, y1 = x_1[i], y_1[i]
    x2, y2 = x_1[i+1], y_1[i+1]

    # Gerade 1: y = m1 * (x - x1) + y1
    # Gerade 2: y = m2 * (x - x2) + y2
    # -> Schnittpunkt bestimmen:
    schnitt_x = (m1 * x1 - m2 * x2 + y2 - y1) / (m1 - m2)
    schnitt_y = m1 * (schnitt_x - x1) + y1

    schnittpunkte_x.append(schnitt_x)
    schnittpunkte_y.append(schnitt_y)

# Plot der Datenpunkte
plt.scatter(x_1, y_1, color='blue', label="Datenpunkte")

# Plot der lokal angepassten Geraden
for i in range(n-1):
    x_start, y_start = x_1[i], y_1[i]
    x_end, y_end = schnittpunkte_x[i], schnittpunkte_y[i]
    
    plt.plot([x_start, x_end], [y_start, y_end], color='orange')

# Achsentitel
plt.title('Kantige Ausgleichskurve für Drehwinkel')
plt.xlabel('Wellenlänge [nm]')
plt.ylabel('Drehwinkel [°]')

# Achsenticks
plt.xticks(np.arange(400, 850, 50))
plt.yticks(np.arange(47, 55, 1))

# Gitter aktivieren
plt.grid(True, linestyle=':', linewidth=0.6)

plt.legend()
plt.show()
```
